## Data Cleaning

We must eliminate rows containing missing values (NaNs) from your dataset. The exchange rate data for the shekel against the euro is incomplete before January 2000. This is because the euro wasn't introduced for non-cash transactions until 1999. Before then, the exchange rate was calculated against other European currencies, like the German mark or French franc. Once the eurozone was established, the shekel-to-euro exchange rate became significant, and data tracking began.

From the above observation, we need to do the below data cleaning activities.

* The data type of 'Period\Unit:' column is string and we need to convert it into datetime.
* Rename the headers such as 'Period\Unit:' and 'Israeli shekel'.
* There are NaN rows in the dataset and we can isolate them to get accurate result.


```python
exchange_rates.rename(columns={'[Israeli shekel ]': 'Israel_shekel',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
```

```python
#Isolate the 'Time' and the 'Israel_shekel' columns. Assign them to a different variable:
euro_to_shekel = exchange_rates[['Time', 'Israel_shekel']]
print(euro_to_shekel)
```

```python
# Isolate the rows with value '-' and 'NaN' in the euro to shekel dataframe.
euro_to_shekel = euro_to_shekel[(euro_to_shekel['Israel_shekel'] != '-') & (euro_to_shekel['Israel_shekel'].notnull())]
euro_to_shekel.reset_index(drop=True, inplace=True)
print(euro_to_shekel)
```

```python
# Convert the column 'Israel_shekel' to float.
euro_to_shekel['Israel_shekel'] = euro_to_shekel['Israel_shekel'].astype(float)

# Examine the 'INR' conversion.
euro_to_shekel.info()
```



---


## Exploratory Data Analysis
We aim to analyze the fluctuations of the Israeli shekel against the euro over time. To achieve this, we can present two types of graphs:

* Short-term Trends: A visual representation of daily changes in the exchange rate.
* Long-term Trends: A visual representation of the rolling average (moving average) of the exchange rate.

```python
import matplotlib.pyplot as plt

plt.plot(euro_to_shekel['Time'], euro_to_shekel['Israel_shekel'])
plt.xlabel('Year')
plt.ylabel('Israeli shekel rate againt euro')
plt.title('Short-term trends')
plt.show()
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_shekel['Time'], euro_to_shekel['Israel_shekel'])
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2, 3, 4, 5, 6],
                           [7, 30, 50, 100, 365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_shekel['Time'],
             euro_to_shekel['Israel_shekel'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')

plt.tight_layout() # Auto-adjusts the padding between subplots
plt.show()
```

## Visualise the relationship between EUR and NIS (New Israeli Shekel)

In this step we will finally get to see the evolution of NIS-EUR exchange rate during specific historical crisis times over the past 25 years. The following events are of our interest:
* The Al-Aqsa Intifada
* Second Lebanon War
* Operation `'Cast Lead'` (Israeli military operation against Hamas in the Gaza Strip)
* Operation `'Pillar of Defense'` (Israeli operation in response to rocket fire from Gaza by Hamas)
* Operation `'Protective Edge'` (A 50-day conflict between Israel and Hamas)
* Operation `'Iron Swords'` (The Israeli response to Hamas' terrorist attack on October 7)

```python
# Function to customize axes
def customize_axes(ax):
    ax.tick_params(left=False)
    for side in ['left', 'right', 'top', 'bottom']:
        ax.spines[side].set_visible(False)
    ax.set_xticklabels([])
    ax.set_xticks([])

# Function to add red dotted lines for key dates and markers
def add_event_lines(ax, dates, data, color='#de0d0d'):
    for date in dates:
        ax.axvline(x=date, color=color, linestyle=':', linewidth=1.5, alpha=0.5)
        red_dot_y_vals = data.loc[data['Time'] == date, 'Israel_shekel'].iloc[0]
        ax.plot([date], red_dot_y_vals, marker='o', color=color, markersize=4)

# Function to adjust x-axis annotations
def annotate_x_axis(ax, dates):
    secax = ax.secondary_xaxis('top')
    secax.set_xticks([dates[-1]])
    secax.set_xticklabels([dates[-1].strftime('%d.%m.%Y')])
    secax.spines['top'].set_visible(False)

# Create a figure with 6 subplots stacked vertically
fig, axes = plt.subplots(nrows=6, ncols=1, figsize=(5, 20))
fig.subplots_adjust(hspace=0.7)

# Customize each axis
for ax in axes:
    customize_axes(ax)

# Adding the title for the visualization
plt.text(0.08, 0.95, 'Six major events in Israel over the past 25 years', size=15, weight='bold', transform=fig.transFigure)
plt.text(0.08, 0.94, 'and their impact on the NIS-EUR exchange rate', size=12, transform=fig.transFigure)

# The Al-Aqsa Intifada (multi-year event)
al_ax = euro_to_shekel[(euro_to_shekel['Time'] >= pd.to_datetime('2000-01-01')) & 
                       (euro_to_shekel['Time'] < pd.to_datetime('2006-01-01'))]
axes[0].plot(al_ax['Time'], al_ax['Israel_shekel'].rolling(30).mean(), color='#757575', linewidth=1.5)
axes[0].set_ylim(2.8, 6.5)

# Add red dotted lines for key dates and annotations
datetimes1 = pd.to_datetime(["2000-09-28", "2005-02-28"])
add_event_lines(axes[0], datetimes1, euro_to_shekel)
annotate_x_axis(axes[0], datetimes1)

# Event details for plotting
events = [
    {"name": "Second Lebanon War", "dates": ["2006-07-12", "2006-08-14"], "time_range": ["2006-07-02", "2006-08-24"]},
    {"name": "Operation Cast Lead", "dates": ["2008-12-24", "2009-01-19"], "time_range": ["2008-12-14", "2009-01-29"]},
    {"name": "Operation Pillar of Defense", "dates": ["2012-11-14", "2012-11-21"], "time_range": ["2012-11-04", "2012-12-01"]},
    {"name": "Operation Protective Edge", "dates": ["2014-07-08", "2014-08-26"], "time_range": ["2014-06-28", "2014-09-05"]},
    {"name": "Operation Iron Swords", "dates": ["2023-10-09"], "time_range": ["2023-09-29", pd.Timestamp.today().strftime('%Y-%m-%d')]}
]

# Loop through events and plot
for i, event in enumerate(events):
    event_data = euro_to_shekel[(euro_to_shekel['Time'] >= event['time_range'][0]) & (euro_to_shekel['Time'] < event['time_range'][1])]
    axes[i + 1].plot(event_data['Time'], event_data['Israel_shekel'], color='#757575', linewidth=1.5)

    # Add red lines and annotations
    event_dates = pd.to_datetime(event['dates'])
    add_event_lines(axes[i + 1], event_dates, euro_to_shekel)
    annotate_x_axis(axes[i + 1], event_dates)

# Final plot adjustments
axes[-1].set_ylim(3.8, 4.4)

# Adding text explanations and descriptions for each event
plt.text(1.1, 0.88, 'The Al-Aqsa Intifada', size=11, weight='bold', transform=fig.transFigure)
plt.text(1.1, 0.87, 'September 28, 2000 - February 28, 2005', fontstyle='italic', transform=fig.transFigure)
plt.text(1.1, 0.858,
         "Description: The Al-Aqsa, which began in September 2000, led to instability in \n"
         "the region and put pressure on Israel's economy. The political and military \n"
         "tensions negatively affected foreign investments and currency exchange rates.",
         size=10, transform=fig.transFigure, verticalalignment='top')
plt.text(1.1, 0.82,
         'Result: During this period, the shekel weakened against \n'
         'the euro due to the increased risk for investors.',
         size=10, transform=fig.transFigure, verticalalignment='top')

plt.text(1.1, 0.75, 'The Second Lebanon War', size=11, weight='bold', transform=fig.transFigure)
plt.text(1.1, 0.74, 'July 12, 2006 - August 14, 2006', fontstyle='italic', transform=fig.transFigure)
plt.text(1.1, 0.728,
         "Description: In July 2006, the Second Lebanon War began between Israel and Hezbollah, \n"
         "the Lebanese Shiite group supported by Iran. The conflict lasted 34 days and \n"
         "resulted in a significant number of casualties and destruction in northern Israel.",
         size=10, transform=fig.transFigure, verticalalignment='top')
plt.text(1.1, 0.69,
         "Result: During the Second Lebanon War, the exchange rate of the shekel to the euro \n"
         "did not experience significant fluctuations. The values remained within a \n"
         "range of 5.5869 to 5.7981, indicating relative stability.",
         size=10, transform=fig.transFigure, verticalalignment='top')

plt.text(1.1, 0.61, "Operation 'Cast Lead'", size=11, weight='bold', transform=fig.transFigure)
plt.text(1.1, 0.60, 'December 24, 2008 - January 19, 2009', fontstyle='italic', transform=fig.transFigure)
plt.text(1.1, 0.588,
         "Description: Event: From December 2008 to January 2009, Israel conducted \n"
         "a military operation against Hamas in response to rocket fire from the Gaza Strip. \n"
         "Operation 'Cast Lead' lasted three weeks.",
         size=10, transform=fig.transFigure, verticalalignment='top')
plt.text(1.1, 0.55,
         "Result: Despite the ongoing military operation, the shekel demonstrated \n"
         "a general trend of depreciation against the euro, indicating economic pressures during this period.",
         size=10, transform=fig.transFigure, verticalalignment='top')

plt.text(1.1, 0.47, "Operation 'Pillar of Defense'", size=11, weight='bold', transform=fig.transFigure)
plt.text(1.1, 0.46, 'November 14, 2012 - November 21, 2012', fontstyle='italic', transform=fig.transFigure)
plt.text(1.1, 0.448,
         "Description: Operation Pillar of Defense, launched in November 2012, was a military response \n"
         "by Israel to escalating rocket fire from Gaza targeting Israeli civilians. The operation aimed \n"
         "to weaken Hamas's rocket-launching capabilities and restore security to southern Israel.",
         size=10, transform=fig.transFigure, verticalalignment='top')
plt.text(1.1, 0.408,
         "Result: During Operation Pillar of Defense, the exchange rate of the shekel to the euro \n"
         "fluctuated slightly, indicating stability. The Israeli economy remained resilient during the conflict.",
         size=10, transform=fig.transFigure, verticalalignment='top')

plt.text(1.1, 0.33, "Operation 'Protective Edge'", size=11, weight='bold', transform=fig.transFigure)
plt.text(1.1, 0.32, 'July 08, 2014 - August 26, 2014', fontstyle='italic', transform=fig.transFigure)
plt.text(1.1, 0.308,
         "Description: Operation 'Protective Edge', launched in July 2014, was Israel's response to increased \n"
         "rocket fire from Gaza. The operation lasted for 50 days and was one of the most significant \n"
         "military conflicts between Israel and Hamas.",
         size=10, transform=fig.transFigure, verticalalignment='top')
plt.text(1.1, 0.268,
         "Result: The shekel weakened against the euro during the operation, with a significant \n"
         "decline in the exchange rate reflecting the economic pressures and risks perceived \n"
         "by investors during the conflict.",
         size=10, transform=fig.transFigure, verticalalignment='top')

plt.text(1.1, 0.18, "Operation 'Iron Swords'", size=11, weight='bold', transform=fig.transFigure)
plt.text(1.1, 0.17, 'October 09, 2023', fontstyle='italic', transform=fig.transFigure)
plt.text(1.1, 0.158,
         "Description: Operation Iron Swords was launched in October 2023 as a defensive \n"
         "military response by Israel to escalating threats in the region.",
         size=10, transform=fig.transFigure, verticalalignment='top')
plt.text(1.1, 0.125,
         "Result: Although the shekel exhibited slight fluctuations, the overall trend \n"
         "indicated stability as the shekel remained within a stable range relative to the euro.",
         size=10, transform=fig.transFigure, verticalalignment='top')

plt.show()
```

The analysis of the shekel to euro exchange rate during six key events of the Israel-Hamas conflict shows that the shekel depreciated in proportion to the duration of military actions.
